# Beautifulsoup

* html 분석을 위한 파이썬 라이브러리
* =beautifulsoup(html코드, html 번역선생님)

```python
pip install bs4
```

```python
import requests 
from bs4 import BeautifulSoup

# naver 서버에 대화를 시도
response = requests.get("http://www.naver.com")

# naver에서 html을 줌
html=response.text

# html 번역 선생님으로 수프 만듦
soup= BeautifulSoup(html, 'html.parser')
#select: 여러개의 태그를 선택하고 싶을 떄
#select_one: 내가 원하는 태그 한개만 선택하고 싶을 떄

# id값이 NM_set_home_bt라고 써있는 한개만 찾아냄
word= soup.select_one("#NM_set_home_btn") 
# NM_set_home_bt라고 써있는id 값을 확인
# id값 앞에 # 를 넣어줘야함. CSS선택자
```

```python
#텍스트 요소만 출력하기
print(word.text)
```

```python
print(word)
```

# CSS 선택자

* CSS란?
 - 웹사이트의 디자인을 표시하기 위한 언어
 - 글자 색 변경, 폰트크기, 가로세로 길이 변경




* CSS 선택자: 디자인을 변경한 HTML 태그를 선택하는 것 (=크롤링할 HTML 태그를 선택하는 것)


* CSS 선택자 종류
  - 태그 선택자: 태그의 이름으로 선택 
  - id 선택자: id 값으로 선택 '#id값'으로 씀
  - class 선택자: clss값으로 선택 '.clss값'으로 씀
  - 자식 선택자: 보통 내가 원하는 태그에 별명이 없을 떄 사용.  '부모태그선택자값>자식태그선택자'으로 씀
